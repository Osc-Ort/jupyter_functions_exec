```python
def funcion_con_closure():
    """Función que retorna una función interna"""
    contador = 0
    
    def incrementar():
        nonlocal contador
        contador += 1
        return contador
    
    return incrementar
```

```python
def funcion_multilinea(
        param1,
        param2,
        param3="default"):
    """Función con parámetros en múltiples líneas"""
    return f"{param1}-{param2}-{param3}"
```
